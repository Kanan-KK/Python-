```python

for i in range(3,1001):
    if i%2==0:
        continue
    elif i%3==0:
       continue
    elif i%5==0:
       continue
    elif i%7==0:
       continue
    elif int(i%(i**0.5))==0:
      continue
    else:
       print(i)
   
```

```python
#FROM 1 TO 1000
for i in range(2,1001):
    simple_number=True
    for j in range(2, int(i**0.5)+1):
        if i % j==0:
            simple_number=False
            break
        if simple_number:
            print(i)
        
```

```python
#MIRROR NUMBER
x=123454321
if str(x)==str(x)[::-1]:
    print('it is palendrom')
else:
    print('not polendrom')
```

```python
#ARMSTRONG NUMBER
x=352
pover=len(str(x))
sum_cubes=sum(int(num)**pover for num in str(x))
print(sum_cubes)
if sum_cubes==x:
    print("Armstrong Numbers")
else:
    print('Not Armstrong Numbers')
```

```python
#UPPER TRIANGLE
row_len=10
col_len=10
for i in range(row_len):
    for j in range(col_len):
        if i<j:
            print('1 ',end= "")
        else:
            print('0 ',end= "")
    print()
```

```python
#LOWER TRIANGLE
row_len=10
col_len=10
for i in range(row_len):
    for j in range(col_len):
        if i>j:
            print('1 ',end= "")
        else:
            print('0 ',end= "")
    print()
```

```python
#TREE
x=15
for i in range(1,x+1):
    space=" " * (x-i)
    star="*" * (2*i-1)
    
    print(space+"\033[32m"+star)
    

    
```
